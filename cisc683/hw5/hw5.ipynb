```python
import pandas as pd

df = pd.read_csv('./hw5-1.csv', dtype={'VET':'category', 'AREA':'category', 'EDUCATION':'category', 'DECISION':'category'})
df
```

```python
def cond_prob(df, att_value, class_value):
    att_values = df[att_value].unique()
    counts = df.groupby(att_value)[class_value].value_counts() + 1
    totals = counts.groupby(level=0).sum()
    return counts/totals

class_value = 'DECISION'
cond_prob(df, 'AREA', class_value)
```

```python
#print (df['AREA'].unique())
#print (df['DECISION'].unique())
#print(ind)
def bayes_prob(df, class_val, att_val):
    ind = pd.MultiIndex.from_product([df[class_val].unique(), df[att_val].unique()], names=(class_val, att_val))

    counts = df.groupby(class_val)[att_val].value_counts()
    #counts = counts.index.isin(vals, level=1)
    #print(counts)
    #print(counts.index)
    counts = counts.reindex(ind, fill_value=0)
    #print(counts)
    counts = counts + 1
    #print(counts)
    totals = counts.groupby(level=0).sum()
    #print(totals)

    probs = counts/totals
    frame = { 'counts': counts, 'probs': probs}
    return pd.DataFrame(frame)

bayes_prob(df, 'DECISION', 'AREA')
```

```python
import math

def cond_prob(x, mu, std):
    z2 = math.pow(((x-mu)/std),2)
    return math.pow(math.e, (-1/2)*z2)/(math.sqrt(2*math.pi) * std)

def row_prob(row):
    x = 27
    mu = row['mean']
    std = row['std']
    return cond_prob(x, mu, std)
    
def num_prob(df, value, attval, classval):
    stats = df.groupby([classval]).agg({attval:['mean','std']})
    stats.columns = ['mean','std']
    return stats

stats = num_prob(df, 27, 'YEARS', 'DECISION')
print(stats)
stats.apply(row_prob, axis=1)
```
