```python
import pandas as pd
import numpy as np

dfbase = pd.read_csv('./hw10-q1-data.csv')
dfbase
```

```python
raw_cols = ['mpg','cylinders', 'horsepower', 'weight', 'acceleration', 'model-year']
raw_cols_all = raw_cols + ['car-name']
norm_cols = ['mpg_n','cylinders_n', 'horsepower_n', 'weight_n', 'acceleration_n', 'model-year_n']
norm_cols_all = norm_cols + ['car-name']

def normalize(df, dfbase):
    for col in raw_cols:
        df[col+'_n'] = (df[col]-dfbase[col].min())/(dfbase[col].max()-dfbase[col].min())
    return df

df = normalize(dfbase, dfbase)
df[norm_cols_all]
```

```python
# test set
df_test = pd.DataFrame(
    [[13,8,150,4464,12,2009,'dodge',0,0,0,0,0,0],
     [20,6,122,2807,13.5,2009,'nissan',0,0,0,0,0,0]],
     columns=dfbase.columns)

df_test = normalize(df_test, dfbase)
df_test[norm_cols_all]
```

```python
def distance(row, df, dcol, mpg=False):
    df[dcol] = 0
    cols = ['cylinders_n', 'horsepower_n', 'weight_n', 'acceleration_n', 'model-year_n']
    if mpg:
        cols = norm_cols
    print(cols)
    for col in cols:
        df[dcol] += (row[col] - df[col]) ** 2
    df[dcol] += (row['car-name'] != df['car-name']) * 1
    df[dcol] = np.sqrt(df[dcol])
```

```python
# compute distances
for idx, row in df_test.iterrows():
    dist_idx = 'dist' + str(idx)
    distance(row, df, dist_idx)
#    df[dist_idx] = 0
#    for col in ['cylinders_n', 'horsepower_n', 'weight_n', 'acceleration_n', 'model-year_n']:
#        df[dist_idx] += (row[col] - df[col]) ** 2
#    df[dist_idx] += (df['car-name'] != row['car-name']) * 1
#    df[dist_idx] = np.sqrt(df[dist_idx])

df[['mpg'] + norm_cols_all + ['dist0', 'dist1']]
```

```python
# k means
df = df.drop(columns=['dist0','dist1'])
centroids = df.loc[[0,5,9]].reset_index()
centroids = centroids.drop(columns=['index', 'mpg', 'cylinders', 'horsepower', 'weight', 'acceleration', 'model-year'])
centroids
```

```python
# assign clusters
def assign_clusters(centroids, df):
    # compute distances
    for cidx, crow in centroids.iterrows():
        distance(crow, df, 'c' + str(cidx), True)
    # compute min distance
    df['cmin'] = df.loc[:, ['c0', 'c1', 'c2']].min(axis=1)
    # assign clusters
    df['cluster'] = 0
    df.loc[df.c0 == df.cmin, 'cluster'] = 0
    df.loc[df.c1 == df.cmin, 'cluster'] = 1
    df.loc[df.c2 == df.cmin, 'cluster'] = 2  
    return df[['mpg_n', 'cylinders_n', 'horsepower_n', 'weight_n', 'acceleration_n', 'model-year_n', 'car-name', 
               'c0', 'c1', 'c2', 'cluster']]
        
df = assign_clusters(centroids, df)
df
```

```python
def find_centroids(df):
    cdf = df.groupby('cluster')['mpg_n', 'cylinders_n', 'horsepower_n', 'weight_n', 'acceleration_n', 'model-year_n'].mean()
    cdf['car-name'] = ''
    cdf.loc[0, 'car-name'] = df[df['cluster']==0]['car-name'].mode()[0]
    cdf.loc[1, 'car-name'] = df[df['cluster']==1]['car-name'].mode()[0]
    cdf.loc[2, 'car-name'] = df[df['cluster']==2]['car-name'].mode()[0]
    return cdf

cdf = find_centroids(df)
cdf
#df[df['cluster']==0]['car-name'].mode()[0]
```
